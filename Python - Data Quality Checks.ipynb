# Data Quality Checks – Logistics Dataset

This notebook performs data quality validation on multiple tables used in the logistics profitability analysis.

Tables included:
- Fact tables: loads, trips, delivery_events
- Cost tables: fuel_purchases, maintenance_records, safety_incidents
- Dimension tables: drivers, trucks, trailers, customers, facilities, routes

The analysis focuses on the following tables that are relevant to the business case:

- Trips
- Loads
- Routes
- Fuel_Purchases
- Maintenance_Records
  

## Import Libraries


```python
from pathlib import Path
import pandas as pd
import numpy as np
```

## Load Data

The datasets are stored as CSV files and loaded into pandas DataFrames for validation and cleaning.

```python

folder_path = Path(r"C:\Users\mory_\Downloads\Compressed\Logistic Database")

trips = pd.read_csv(folder_path/'trips.csv')
loads = pd.read_csv(folder_path / "loads.csv")
routes = pd.read_csv(folder_path / "routes.csv")
FUEL_PURCHASES = pd.read_csv(folder_path / "FUEL_PURCHASES.csv")
MAINTENANCE_RECORDS = pd.read_csv(folder_path / "MAINTENANCE_RECORDS.csv")
```

## Initial Data Inspection


```python
## Data Inspection and Data Quality Checks

def quality_check(df, name):
    
    print(f"\n---- {name} ----")
    
    print("Rows:", df.shape[0])
    print("Columns:", df.shape[1])
    
    print("\nMissing values:")
    print(df.isnull().sum())
    
    print("\nDuplicate rows:", df.duplicated().sum())

    print("\nChecking Data Types:")
    print(df.dtypes)

    print("\nChecking Invalid Values and basic Statistics:")
    print(df.describe())
```

```python

quality_check(trips, "Trips Table")
quality_check(loads, "Loads Table")
quality_check(routes, "Routes Table")
quality_check(FUEL_PURCHASES, "FUEL_PURCHASES Table")
quality_check(MAINTENANCE_RECORDS, "MAINTENANCE_RECORDS Table")
```

## Data Cleaning


```python
trips = trips.dropna(subset=["truck_id"])
```

```python
# check rows with null values
trips[trips.isnull().any(axis=1)].head()
```

```python
# missing values for FUEL_PURCHASES
FUEL_PURCHASES = FUEL_PURCHASES.dropna(subset=['truck_id'])
```

```python
FUEL_PURCHASES[FUEL_PURCHASES.isnull().any(axis=1)].head()
```

## Export Cleaned Data
Cleaned datasets were saved for further SQL analysis and Power BI dashboard development.

```python
# Save Clean Data
trips.to_csv(folder_path/"trips_clean.csv", index=False)
FUEL_PURCHASES.to_csv(folder_path/'FUEL_PURCHASES_clean.csv',index=False)
```
